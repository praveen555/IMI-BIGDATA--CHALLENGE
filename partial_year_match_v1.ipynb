```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
def fuzzyscore(col):
  return fuzz.token_sort_ratio(col['NAME'],col['name'])

```

```python
sanction=sanction.reset_index(drop=True)
```

```python
cust_data=cust_data.reset_index(drop=True)
```

```python
cust_name_list,sanction_name_list,year,score=[],[],[],[]
for ind in sanction.index:
  print(sanction['year'][ind],ind)
  x=cust_data[cust_data['year']==sanction['year'][ind]]
  if x.empty!=True:
    x['name']=sanction['name'][ind]
    x['score']=x.apply(fuzzyscore,axis=1)
    x.sort_values(by=['score'],inplace=True,ascending=False)
    x=x[x['score']>=75]
    if x.empty!=True:
      cust_name_list.append(x['NAME'].to_list())
      sanction_name_list.append(x['name'].to_list())
      year.append(x['year'].to_list())
      score.append(x['score'].to_list())
      print("Threshold names found")
    else:
      print("Inner empty df")
      pass
    # print(x)
    # print(cust_name_list,sanction_name_list,year,score)
  else:
    print("Skip empty df")
 
```

```python
print(len(score),len(cust_name_list),len(sanction_name_list))
```

```python
d={'bank_customer_name':cust_name_list,'sanction_name_list':sanction_name_list,'year':year,'score':score}
```

```python
df=pd.DataFrame(data=d)
```
