## Visuzlize data using Kmeans

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
pca_num_components = 2
```

```python
map={'low':0,'medium':1,'high':2}
low['RISK']=low['RISK'].map(map)
low_x=low.drop(['BIRTH_DT','CUST_ADD_DT','NAME','CUSTOMER_ID'],axis=1)
low_x_data=low_x.iloc[:,low_x.columns!='RISK']
low_y=low_x['RISK']
low_x_data=pd.get_dummies(low_x_data)
print(low_x_data.shape,low_y.shape)
kmeans_low= KMeans(n_clusters=3, init='k-means++',max_iter=50000,random_state=0).fit(low_x_data)
reduced_data = PCA(n_components=pca_num_components).fit_transform(low_x_data)
results_low= pd.DataFrame(reduced_data,columns=['pca1','pca2'],index=low.index)
results_low['label']=kmeans_low.labels_
sns.scatterplot(x="pca1", y="pca2", hue=kmeans_low.labels_, data=results_low)
plt.title('K-means Clustering with 2 dimensions for Low Class')
plt.legend()
plt.show()
```

```python
map={'low':0,'medium':1,'high':2}
med['RISK']=med['RISK'].map(map)
med_x=med.drop(['BIRTH_DT','CUST_ADD_DT','NAME','CUSTOMER_ID'],axis=1)
med_x_data=med_x.iloc[:,med_x.columns!='RISK']
med_y=med_x['RISK']
med_x_data=pd.get_dummies(med_x_data)
print(med_x_data.shape,med_y.shape)
kmeans_med= KMeans(n_clusters=3, init='k-means++',max_iter=50000,random_state=0).fit(med_x_data)
reduced_data = PCA(n_components=pca_num_components).fit_transform(med_x_data)
results_med= pd.DataFrame(reduced_data,columns=['pca1','pca2'],index=med.index)
results_med['label']=kmeans_med.labels_
sns.scatterplot(x="pca1", y="pca2", hue=kmeans_med.labels_, data=results_med)
plt.title('K-means Clustering with 2 dimensions for Medium Class')
plt.legend()
plt.show()
```

```python
map={'low':0,'medium':1,'high':2}
high['RISK']=high['RISK'].map(map)
high_x=high.drop(['BIRTH_DT','CUST_ADD_DT','NAME','CUSTOMER_ID'],axis=1)
high_x_data=high_x.iloc[:,high_x.columns!='RISK']
high_y=high_x['RISK']
high_x_data=pd.get_dummies(high_x_data)
print(high_x_data.shape,high_y.shape)
kmeans_high= KMeans(n_clusters=3, init='k-means++',max_iter=50000,random_state=0).fit(high_x_data)
reduced_data = PCA(n_components=pca_num_components).fit_transform(high_x_data)
results_high= pd.DataFrame(reduced_data,columns=['pca1','pca2'],index=high.index)
results_high['label']=kmeans_high.labels_
sns.scatterplot(x="pca1", y="pca2", hue=kmeans_high.labels_, data=results_high)
plt.title('K-means Clustering with 2 dimensions High Class')
plt.legend()
plt.show()
```

## Feature Engineering

```python
for i in c:
  data[i]=data[i]+1
```

```python
data['cash_diff']=data['CASH_SUM_IN']-data['CASH_SUM_OUT'] #difference in cash
data['wire_diff']=data['WIRES_SUM_IN']-data['WIRES_SUM_OUT'] # difference in wire
data['cash_cnt_diff']=data['CASH_CNT_IN']-data['CASH_CNT_OUT'] # difference in cash counts
data['wire_cnt_diff']=data['WIRES_CNT_IN']-data['WIRES_CNT_OUT'] # difference in wire counts
data['avg_cash_transaction_out']=data['CASH_SUM_OUT']/data['CASH_CNT_OUT'] # per cnt cash out value
data['avg_wire_transaction_out']=data['WIRES_SUM_OUT']/data['WIRES_CNT_OUT'] # per cnt wire out value
data['avg_cash_transaction_in']=data['CASH_SUM_IN']/data['CASH_CNT_IN'] # per cnt cash in value
data['avg_wire_transaction_in']=data['WIRES_SUM_IN']/data['WIRES_CNT_IN'] # per cnt cash out value
data['Wcnt_in_per_month']=(data['WIRES_CNT_IN']/12) #cnt per month
data['Wcnt_out_per_month']=(data['WIRES_CNT_OUT']/12)# cnt per month 
data['Ccnt_in_per_month']=(data['CASH_CNT_IN']/12)
data['Ccnt_out_per_month']=(data['CASH_CNT_OUT']/12)
data['monthly_cash_in']=data['CASH_SUM_IN']/12 #estimated monthly cash in
data['monthly_cash_out']=data['CASH_SUM_OUT']/12 # estimated monthly cash out
data['monthly_wire_in']=data['WIRES_SUM_IN']/12 
data['monthly_wire_out']=data['WIRES_SUM_OUT']/12
data['Ccnt_in_per_month']=data['monthly_cash_in']/data['Ccnt_in_per_month'] # estimated cash in per montly cnt
data['Ccnt_out_per_month']=data['monthly_cash_out']/data['Ccnt_out_per_month']
data['Wcnt_in_per_month']=data['monthly_wire_in']/data['Wcnt_in_per_month']
data['Wcnt_out_per_month']=data['monthly_wire_out']/data['Wcnt_out_per_month']
data['diff_year']=data['CUST_ADD_DT'].dt.year-data['BIRTH_DT'].dt.year # age when opened bank account
data['age']=2023-data['BIRTH_DT'].dt.year #age of customer
data['cust_age_bank']=2023-data['CUST_ADD_DT'].dt.year # number of years assocaited with bank
data['total_money_left']=data['CASH_SUM_IN']+data['WIRES_SUM_IN']-data['CASH_SUM_OUT']-data['WIRES_SUM_OUT'] # left out money yearly
data['IN_AVG']=(data['CASH_SUM_IN']+data['WIRES_SUM_IN'])/2 #avg in
data['OUT_AVG']=(data['CASH_SUM_OUT']+data['WIRES_SUM_OUT'])/2 # avg out
data['IN']=(data['CASH_SUM_IN']+data['WIRES_SUM_IN']) # total in 
data['OUT']=(data['CASH_SUM_OUT']+data['WIRES_SUM_OUT']) # total out
```
