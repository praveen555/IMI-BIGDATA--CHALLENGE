```python
def clean_text(text):
  token=str(text).split(";")
  #tokens = [word for word in token if word.isascii()==False]
  #table = str.maketrans("","",punctuation)
  #token = [word.translate(table) for word in token]
  #words=[word.replace("","""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~""") for word in token]
  words=[word.lower() for word in token]
  return words
```

```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
def fuzzyscore(col):
  return fuzz.token_set_ratio(col['NAME'],col['name'])

```

```python

def fuzzyscore_2(col):
  return process.extractOne(col['NAME'],col['aliases'],scorer=fuzz.token_set_ratio)
```

```python
cust_id,cust_name_list,sanction_name_list,list_aliases,birth_dt,score=[],[],[],[],[],[]
aliases_name,aliases_score=[],[]
for ind in sanction.index:
  if ind%1000==0:
    print("Iteration--",ind)
  #print(sanction['year'][ind],ind)
  data=cust_data[cust_data['year']==sanction['year'][ind]]
  if data.empty!=True:
    #print(sanction['name'][ind],sanction['aliases'][ind])
    data['name']=sanction['name'][ind]
    data['aliases']=sanction['aliases'][ind]
    data['aliases']=data['aliases'].astype(str).apply(clean_text)
    data['score']=data.apply(fuzzyscore,axis=1)
    data['aliases_data']=data.apply(fuzzyscore_2,axis=1)
    data['aliases_name']=data['aliases_data'].str[0]
    data['aliases_score']=data['aliases_data'].str[1]
    data=data[['CUSTOMER_ID','NAME','BIRTH_DT','name','aliases','score','aliases_name','aliases_score']]
    #print(data)
    data.sort_values(by=['score'],inplace=True,ascending=False)
    data=data[data['score']>=75]
    #print(data)
    if data.empty!=True:
      cust_id.append(data['CUSTOMER_ID'].to_list())
      cust_name_list.append(data['NAME'].to_list())
      birth_dt.append(data['BIRTH_DT'].to_list())
      sanction_name_list.append(data['name'].to_list())
      list_aliases.append(data['aliases'].to_list())
      score.append(data['score'].to_list())
      aliases_name.append(data['aliases_name'].to_list())
      aliases_score.append(data['aliases_score'].to_list())
      #print("Threshold names found")
    else:
      #print("Inner empty df")
      pass
    # print(x)
    # print(cust_name_list,sanction_name_list,year,score)
  else:
    pass
    #print("Skip empty df")
```
