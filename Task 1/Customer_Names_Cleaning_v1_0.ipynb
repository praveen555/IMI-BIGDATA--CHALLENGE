```python
import pandas as pd
```

```python
data=pd.read_csv('UofT_nodes.csv')
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
df = data[data['NAME'].notna()]
```

```python
df
```

```python
df.isna().sum()
```

```python
df_names=df[['CUSTOMER_ID','NAME','BIRTH_DT']]
```

```python
df_names
```

```python
spec_chars = ["!",'"',"#","%","&","'","(",")",
              "*","+",",","-",".","/",":",";","<",
              "=",">","?","@","[","\\","]","^","_",
              "`","{","|","}","~","–"]
```

```python
for char in spec_chars:
    df_names['NAME'] = df_names['NAME'].str.replace(char, ' ')
```

```python
df_names['NAME'] = df_names['NAME'].str.split().str.join(" ")
```

```python
df_names['NAME'] = df_names['NAME'].str.lower()
```

```python
df_names
```

```python
df_names['length']=df_names['NAME'].str.len()
```

```python
df_names.head()
```

```python
df_final_names=df_names[df_names['length']>2]
```

```python
df_final_names=df_final_names[['CUSTOMER_ID','NAME','BIRTH_DT']]
```

```python
df_final_names['CUSTOMER_ID']=df_final_names['CUSTOMER_ID'].astype(int)
```

```python
df_final_names
```

```python
df_final_names.to_csv('ScotiaBank_names.csv')
```
