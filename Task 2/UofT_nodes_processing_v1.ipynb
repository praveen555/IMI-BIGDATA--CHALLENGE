```python
from datetime import datetime, timedelta
from dateutil.relativedelta import *
```

```python
date= datetime(2000,1,1)
date = date + relativedelta(years=+33,months=+6,days=+10)
print(date.date())
```

```python
def filldates_high_medium(dob_date):
  cust_add_dt=dob_date+relativedelta(years=+35,months=+4,days=+10)
  return cust_add_dt.date()
```

```python
def filldates_low(dob_date):
  cust_add_dt=dob_date+relativedelta(years=+31,months=+4,days=+10)
  return cust_add_dt.date()
```

```python
high_medium=data[(data['RISK']=='high') | (data['RISK']=='medium')]
```

```python
idx_high_medium=high_medium.index
```

```python
low=data[(data['RISK']=='low')]
idx_low=low.index
```

```python
df_nodes.loc[idx_high_medium, 'CUST_ADD_DT'] = df_nodes.loc[idx_high_medium, 'BIRTH_DT'].apply(filldates_high_medium)
```

```python
df_nodes.loc[idx_low, 'CUST_ADD_DT'] = df_nodes.loc[idx_low, 'BIRTH_DT'].apply(filldates_low)
```

## Mapping categorical columns 

Since this more based on ranking system such as high,medium and low. Ordinal encoding could be done instead of one hot encoding as there exists a relationship.

```python
df_nodes.columns
```

```python
df_nodes['COUNTRY_RISK_RESIDENCY'].value_counts()
```

```python
mapper1={'low':0,'medium':1,'high':2}
mapper2={'Low':0,'Moderate':1,'High':2}
```

```python
df_nodes['RISK']=df_nodes['RISK'].map(mapper1)
```

```python
df_nodes['COUNTRY_RISK_RESIDENCY']=df_nodes['COUNTRY_RISK_RESIDENCY'].map(mapper2)
df_nodes['OCPTN_NM']=df_nodes['OCPTN_NM'].map(mapper2)
```

```python
df_nodes['COUNTRY_RISK_INCOME']=df_nodes['COUNTRY_RISK_INCOME'].map(mapper2)
```
