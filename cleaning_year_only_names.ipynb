```python
import pandas as pd
```

```python
df=pd.read_csv('test_year_names.csv')
```

```python
df
```

```python
df.drop(['Unnamed: 0','DOB'],inplace=True,axis=1)
```

```python
df
```

```python
df.columns
```

```python
from ast import literal_eval

df['customer_id']=df['customer_id'].apply(literal_eval) #convert to list type
df['bank_customer_name']=df['bank_customer_name'].apply(literal_eval)
df['sanction_name_list']=df['sanction_name_list'].apply(literal_eval)
df['aliases']=df['aliases'].apply(literal_eval)
#df['DOB']=df['DOB'].apply(literal_eval)
df['name_score']=df['name_score'].apply(literal_eval)
df['selected_alias']=df['selected_alias'].apply(literal_eval)
df['aliases_score']=df['aliases_score'].apply(literal_eval)
```

```python
df.shape
```

```python
df=df.explode(['customer_id', 'bank_customer_name', 'sanction_name_list', 'aliases', 'name_score', 'selected_alias', 'aliases_score'])
```

```python
df.shape
```

```python
df
```

```python
df['name_score']=df['name_score'].astype(int)
```

```python
df['aliases_score']=df['aliases_score'].astype(int)
```

```python
df=df[df['name_score']>=95]
```

```python
df
```

try to reduce false positives 

```python
df['cust_word']=df['bank_customer_name'].str.split(" ").str.len()
```

```python
df['sanct_words']=df['sanction_name_list'].str.split(" ").str.len()
```

```python
df[df['cust_word']==df['sanct_words']]
```

```python
df['diff']=df['cust_word']-df['sanct_words']
```

```python
df=df[(df['diff']==1)|(df['diff']==0)]
```
