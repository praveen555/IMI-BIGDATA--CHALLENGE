```python
from string import punctuation
def clean_text(text):
  token=str(text).split(";")
  #tokens = [word for word in token if word.isascii()==False]
  #table = str.maketrans("","",punctuation)
  #token = [word.translate(table) for word in token]
  #words=[word.replace("","""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~""") for word in token]
  words=[word.lower() for word in token]
  return words
```

```python
cust_name=pd.read_csv('ScotiaBank_names_full.csv')
```

```python
cust_name
```

```python
cust_name=cust_name[['CUSTOMER_ID','NAME','BIRTH_DT','RISK']]
```

```python
df=cust_name.merge(sanction,on=['BIRTH_DT'])
```

```python
df.shape
```

```python
df
```

```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
def fuzzyscore(col):
  return fuzz.token_set_ratio(col['NAME'],col['name'])
```

```python
df['name_score']=df.apply(fuzzyscore,axis=1)
```

```python
def fuzzyscore_2(col):
  return process.extractOne(col['NAME'],col['aliases'],scorer=fuzz.token_set_ratio)
```

```python
df
```

```python
df.head(10).apply(fuzzyscore_2,axis=1)
```

```python
df['aliases_score']=df.apply(fuzzyscore_2,axis=1)
```

```python
df
```

```python
df.sort_values(['name_score'],ascending=False,inplace=True)
```

```python
df
```

```python
df_threshold=df[df['name_score']>=70]
```

```python
df_threshold.shape
```

```python
df_threshold['aliases_name']=df_threshold['aliases_score'].str[0]
df_threshold['aliases_score']=df_threshold['aliases_score'].str[1]
```

```python
df_threshold
```

```python
x=df_threshold[(df_threshold['aliases_score']>=80)]
```

```python
x.sort_values(['aliases_score'],inplace=True,ascending=False)
```

```python
x
```

```python
x.columns
```

```python
x=x[['CUSTOMER_ID', 'NAME', 'BIRTH_DT','name', 'aliases', 'name_score', 'aliases_score', 'aliases_name']]
```

# Let us try to reduce false positive from this list

1. On a general observation all those aliases score where score is 100 by the match aliases is only a single word we should reject those as they are highly unlikely since only one word from the entire name is matching. However there should be check if the customer name is also one. 

```python
x['aliases_len']=x['aliases_name'].str.split(" ").str.len()
```

```python
x['name_len']=x['NAME'].str.split(" ").str.len()
```

```python
x
```

```python
x=x[~(x['aliases_len']==1) & (x['name_len']!=1)]
```

```python
x=x[x['aliases_score']>=85]
```
