```python
from string import punctuation
def clean_text(text):
  token=str(text).split(";")
  #tokens = [word for word in token if word.isascii()==False]
  #table = str.maketrans("","",punctuation)
  #token = [word.translate(table) for word in token]
  #words=[word.replace("","""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~""") for word in token]
  words=[word.lower() for word in token]
  return words
```

```python
sanction['aliases']=sanction['aliases'].apply(clean_text) 
```

```python
cust_name=pd.read_csv('ScotiaBank_names_v1.csv')
```

```python
cust_name
```

```python
df=cust_name.merge(sanction,on=['BIRTH_DT'])
```

```python
df.shape
```

```python
df
```

```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
def fuzzyscore(col):
  return fuzz.token_set_ratio(col['NAME'],col['name'])
```

```python
df['name_score']=df.apply(fuzzyscore,axis=1)
```

```python
def fuzzyscore_2(col):
  return process.extractOne(col['NAME'],col['aliases'],scorer=fuzz.token_set_ratio)
```

```python
df
```

```python
df.head(10).apply(fuzzyscore_2,axis=1)
```

```python
df['aliases_score']=df.apply(fuzzyscore_2,axis=1)
```

```python
df
```

```python
df.sort_values(['name_score'],ascending=False,inplace=True)
```

```python
df
```

```python
df_threshold=df[df['name_score']>=70]
```

```python
df_threshold.shape
```

```python
df_threshold['aliases_name']=df_threshold['aliases_score'].str[0]
df_threshold['aliases_score']=df_threshold['aliases_score'].str[1]
```

```python
df_threshold
```

```python
x=df_threshold[(df_threshold['aliases_score']>=80)]
```

```python
x.sort_values(['aliases_score'],inplace=True,ascending=False)
```
