```python
def clean_text(text):
  token=str(text).split(";")
  #tokens = [word for word in token if word.isascii()==False]
  #table = str.maketrans("","",punctuation)
  #token = [word.translate(table) for word in token]
  #words=[word.replace("","""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~""") for word in token]
  words=[word.lower() for word in token]
  return words
```

```python
df['aliases']=df['aliases'].apply(clean_text)
```

```python
df
```

```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
def fuzzyscore(col):
  return fuzz.token_set_ratio(col['NAME'],col['name'])

```

```python
df['name_score']=df.apply(fuzzyscore,axis=1)
```

```python
df
```

```python

def fuzzyscore_2(col):
  return process.extractOne(col['NAME'],col['aliases'],scorer=fuzz.token_set_ratio)
```

```python
df.head(10).apply(fuzzyscore_2,axis=1)
```

```python
df['aliases_score']=df.apply(fuzzyscore_2,axis=1)
```

```python
df.sort_values(['name_score'],ascending=False,inplace=True)
```

```python
df
```

```python
df_threshold=df[df['name_score']>=80]
```
